```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import lxml
```

```python
url = "https://endoflife.date/windows"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extract all tables from the page
tables = soup.find_all('table')
print(f"Found {len(tables)} table(s)\n")

# Extract data from each table
all_data = []
for i, table in enumerate(tables):
    rows = table.select('tr, tr.release, tr.release.can-be-hidden')
    table_data = []
    for row in rows:
        cols = row.find_all(['td', 'th'])
        row_data = []
        for col in cols:
            text = col.get_text(separator=' ', strip=True)
            text = ' '.join(text.replace('\n', ' ').replace('\r', ' ').split())
            row_data.append(text)
        if row_data and len(row_data) > 1:
            table_data.append(row_data)
    
    if table_data:
        # Convert to DataFrame for better visualization
        df = pd.DataFrame(table_data[1:], columns=table_data[0])
        all_data.append(df)
        print(f"Table {i+1}:")
        df = pd.concat(all_data, ignore_index=True, sort=False)
       
# Store all dataframes in a dictionary for easy access
tables_dict = {f"table_{i+1}": df for i, df in enumerate(all_data)}


df.head()
```

```python
# Combine all extracted tables and save to CSV
if all_data:
    combined_df = pd.concat(all_data, ignore_index=True, sort=False)
    combined_df.to_csv('windows.csv', index=False)
    print('Saved combined table data to windows.csv')
    display(combined_df.head())
else:
    print('No table data found to save.')
```
